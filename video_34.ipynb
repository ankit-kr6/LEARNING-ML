```python
import zipfile
with zipfile.ZipFile('orders-csv.zip', 'r') as zip_ref:
    zip_ref.extractall('/content')
```

```python
import pandas as pd
import numpy as np
```

```python
date=pd.read_csv('orders.csv')
```

```python
date.head()
```

```python
date['order_date']=pd.to_datetime(date['order_date'])
```

```python
date['date_year']=date['order_date'].dt.year
date.head()
```

```python
date['date_month']=date['order_date'].dt.month
date.head()
```

```python
date['date_month_name']=date['order_date'].dt.month_name()
date.head()
```

```python
#day of week
date['date_day_of_week']=date['order_date'].dt.dayofweek
date.head()
```

```python
#day of week - name
date['date_day_of_week_name']=date['order_date'].dt.day_name()
date.head()
```

```python
#is weekend
date['is_weekend']=np.where(date['date_day_of_week_name'].isin(['Sunday','Saturday']),1,0)
date.head()
```

```python
#extract week of the year
date['date_week'] = date['order_date'].dt.isocalendar().week
date.head()
```
