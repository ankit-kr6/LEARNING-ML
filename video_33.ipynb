```python
import zipfile
with zipfile.ZipFile('titanicdataset-traincsv.zip', 'r') as zip_ref:
    zip_ref.extractall('/content')
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('train.csv',usecols=["Cabin","Ticket","Survived","PassengerId"])
```

```python
df.head()
```

```python
df['Cabin_num']=df['Cabin'].str.extract('(\d+)')
df['cabin_cat']=df['Cabin'].str[0]
df.head()
```

```python
df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
df['ticket_num']=df['Ticket'].apply(lambda s:s.split()[-1])
df['ticket_num']=pd.to_numeric(df['ticket_num'],errors='coerce',downcast='float')
df['ticket_cat']=df['Ticket'].apply(lambda s:s.split()[0])
df['ticket_cat']=np.where(df['ticket_cat'].str.isdigit(),df['Ticket'].str[0],df['ticket_cat'])
df.head()
```
